```python
import networkx as nx
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
from queue import PriorityQueue
```

```python
G= nx.random_geometric_graph(100,0.18)     # generate a random graph
G = nx.to_directed(G)                       # make the graph directed
```

```python
#nx.draw_random(G)
```

```python
edge_x = []                                 # append the edge coordinates to an empty list
edge_y = []
for edge in G.edges():
    x0,y0 = G.nodes[edge[0]]['pos']
    x1,y1 = G.nodes[edge[1]]['pos']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_y.append(y0)
    edge_y.append(y1)
```

```python
edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=0.25, color='#010103'), mode='lines')
```

```python
node_x = []                                # append the node coordinates to an empty list
node_y =[]
label=0
for node in G.nodes():
    x = G.nodes[node]['pos']
    y= G.nodes[node]['pos']
   
    
    node_x.append(x)
    node_y.append(y)
```

```python
node_trace = go.Scatter( x = node_x, y= node_y,ids=np.arange(200), mode='markers', marker=dict(
        showscale=True,
        colorscale='RdBu',
        reversescale=True,
        color=[],
        size=10,
        line_width=2) )
```

```python
Graph = go.Figure(data = [edge_trace, node_trace],layout= go.Layout(
                titlefont_size=16,
                showlegend=False))                   # plot the network graph
```

```python
for e in G.edges():
    G[e[0]][e[1]]['w'] = np.random.randint(1,10)     # add random weight to each edge in G
```

```python
h = PriorityQueue()                      # initiate a priority queue to contain temp. distances
dist= np.ones((len(G.nodes())))          
parent_= [[]]*100                        # 2d list to contain the shortest paths of each node to its source
```

```python
def dijkstra(G, source, target):
    
    for node in G.nodes():
        dist[node] = 1e6                           # assign distance of all nodes to a large number 
        h.put((dist[node],node))                   # add the nodes into the priority queue
    dist[source] = 0                               # set the dist. of source vertex to 0
    h.put((dist[source],source))                   # update the priority queue
    
    while not h.empty():                     
        
        d,u = h.get()                              # extract the min. priority value from h based on min. distance 
        for u,i in G.edges():                      # vertex in h with min. distance
            if dist[u] + G[u][i]['w'] < dist[i]:   
                dist[i] = dist[u] + G[u][i]['w']   # update the distance
                parent_[i] = u                     # kepp track of the parent node
                h.put((dist[i],i))                 # update the priority queue 
                
                if i == target:
                    return print('the shortest distance is', int(dist[target])), path(target,source)
                    break
    print('node not reachable')
   
            

    
```

```python
def path(target,source):
    path_ = []
    path_.append(target)
    i = target
    while True:
        path_.append(parent_[i])
        i = parent_[i]
        if i == source:
            break
    print(path_)
    
```
